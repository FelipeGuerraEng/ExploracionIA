```python
import sys
import csv
sys.path.append('utils')

import blob_client
```

```python
blob_manager = blob_client.BlobFunctions(container_name="bronze")
```

```python
blob_list = blob_manager.listar_archivos_mas_grandes_por_carpeta()
```

```python
len(blob_list)
```

```python
# Nombre del archivo CSV donde se guardará la lista
csv_filename = 'data/blob_list.csv'

# Escribir la lista de blobs en un archivo CSV
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Escribir cada blob en una fila
    for blob in blob_list:
        writer.writerow([blob])
```

```python
# Nombre del archivo CSV donde se guardará la lista
csv_filename = 'data/blob_list.csv'

# Leer el archivo CSV y cargar las rutas de blobs en una lista
blob_list = []
with open(csv_filename, mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        blob_list.append(row[0])  # Cada fila tiene un solo elemento, que es la ruta del blob
```

```python
blob_list
```

```python
# Calcular el tamaño total de los blobs
blobs_size = blob_manager.calcular_tamano_total_blobs(blob_list)
```

```python
# Filtrar la lista para eliminar los elementos None
blobs_size = [size for size in blobs_size if size is not None]

sum(blobs_size)/1e6
```

```python
import matplotlib.pyplot as plt


# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(blobs_size, bins=500, edgecolor='black', alpha=0.7)
plt.title('Histograma de Tamaños de Blobs en Kilobytes')
plt.xlabel('Tamaño (KB)')
plt.ylabel('Frecuencia')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
